**3- Data Cleaning and Visualization**

3.1- Addressing Anomalies

The initial EDA revealed **negative values** in `total_amount` and **zero passenger counts**, which are illogical for standard trip transactions. Negative totals likely represent refunds. Trips with zero passengers are also invalid for ride-share models.


```python
# Filter the data to exclude anomalies:
# 1. Total amount must be positive.
# 2. Fare amount must be positive.
# 3. Passenger count must be greater than zero.

df_clean = df[(df['total_amount'] > 0) & (df['fare_amount'] > 0) & (df['passenger_count'] > 0)]

print(f"Original rows: {len(df)}")
print(f"Cleaned rows: {len(df_clean)}")
print(f"Rows removed: {len(df) - len(df_clean)}")
```

3.2- Payment Type and Tipping Analysis

```python
# How many of each payment type are represented in the cleaned data? (and calculate percentages)
payment_counts = df_clean['payment_type'].value_counts()
payment_percentages = df_clean['payment_type'].value_counts(normalize=True) * 100
payment_summary = pd.DataFrame({'count': payment_counts, 'percentage': payment_percentages.round(2)})
```

```python
# Map numeric payment types to labels for clarity (1=Credit card, 2=Cash, 3=No charge, 4=Dispute)
payment_summary.rename(index={1:'Credit card', 2:'Cash', 3:'No charge', 4:'Dispute'}, inplace=True)
print(payment_summary)
```

```python
# Plotting the payment types
plt.figure(figsize=(6, 3))
sns.barplot(x=payment_summary.index, y=payment_summary['count'], palette='Spectral')
plt.title('Distribution of Trip Counts by Payment Method ðŸ’³', fontsize=14)
plt.xlabel('Payment Type')
plt.ylabel('Number of Trips (Count)')
plt.show()
```

```python
# What is the average tip for trips paid for with credit card vs. cash?
cc_tip = df_clean[df_clean['payment_type']==1]['tip_amount'].mean()
cash_tip = df_clean[df_clean['payment_type']==2]['tip_amount'].mean()

print('Avg. cc tip:', round(cc_tip, 2))
print('Avg. cash tip:', round(cash_tip, 2))
```

```python
# Visualize the average tip amount by payment type
plt.figure(figsize=(6, 4))
tip_means = pd.Series([cc_tip, cash_tip], index=['Credit Card (1)', 'Cash (2)'])
sns.barplot(x=tip_means.index, y=tip_means.values, palette=['#1f77b4', '#ff7f0e'])
plt.title('Average Tip Amount by Payment Type', fontsize=14)
plt.xlabel('Payment Type')
plt.ylabel('Average Tip Amount ($)')
plt.show()
```

```python
# How many times is each vendor ID represented in the data?
df_clean['VendorID'].value_counts()

# What is the mean total amount for each vendor?
df_clean.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

```python
# Filter the data for credit card payments only
credit_card = df_clean[df_clean['payment_type']==1]

# Filter the data for passenger count only (for credit card trips)
cc_passenger_counts = credit_card['passenger_count'].value_counts().sort_index()
print("Credit Card Trips by Passenger Count:\n", cc_passenger_counts)

# Calculate the average tip amount for each passenger count (credit card payments only)
cc_tip_by_passenger = credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']]
print("\nAverage Tip by Passenger Count (Credit Card):\n", cc_tip_by_passenger)
```

```python
# Visualize average tip by passenger count
plt.figure(figsize=(8, 4))
sns.barplot(x=cc_tip_by_passenger.index, y=cc_tip_by_passenger['tip_amount'], palette='coolwarm')
plt.title('Average Credit Card Tip Amount by Passenger Count', fontsize=14)
plt.xlabel('Passenger Count')
plt.ylabel('Average Tip Amount ($)')
plt.show()
```
